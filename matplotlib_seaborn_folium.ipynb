```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import mpl_toolkits.mplot3d.axes3d
%matplotlib inline
```

```python
fig = plt.figure()
p = plt.subplot(111, projection = "3d")

x = np.arange(10) # data x
y = np.arange(10) # data y
z = np.zeros(10) # set tinggi z

dx = np.ones(10) # set ukuran chart x
dy = np.ones(10) # set ukuran chart y
dz = np.arange(10) # data z

p.bar3d(x, y, z, dx, dy, dz)
p.set_xlabel('Sumbu X')
p.set_ylabel('Sumbu Y')
p.set_zlabel('Sumbu Z')
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights.isna().any()
```

```python
sns.relplot(data=flights, x='passengers', y='year', hue='month', size='passengers')
```

```python
flights_year = flights.pivot_table(index='year', columns='month', values='passengers')
sns.heatmap(flights_year)
```

```python
months = flights.month.tolist()
numbers = []
for month in months:
    if month == 'January':
        numbers.append(1)
    elif month == 'February':
        numbers.append(2)
    elif month == 'March':
        numbers.append(3)
    elif month == 'April':
        numbers.append(4)
    elif month == 'May':
        numbers.append(5)
    elif month == 'June':
        numbers.append(6)
    elif month == 'July':
        numbers.append(7)
    elif month == 'August':
        numbers.append(8)
    elif month == 'September':
        numbers.append(9)
    elif month == 'October':
        numbers.append(10)
    elif month == 'November':
        numbers.append(11)
    elif month == 'December':
        numbers.append(12)
        
numbers = pd.Series(numbers, name='month_number')
new_flights = pd.concat([flights, numbers], axis=1)
new_flights
```

```python
correlation_flights = new_flights.pivot_table(index='year', columns=['month','month_number'], values='passengers').corr()
correlation_flights
```

```python
sns.heatmap(correlation_flights)
```

```python
# bisa untuk mmemprediksi data passengers di tahun berikutnya
sns.lmplot(data=flights, x='year', y='passengers', hue='month')
```
